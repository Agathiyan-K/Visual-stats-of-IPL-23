```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the DataFrame from the orange code
# Assuming you already have df_orange loaded

# Define the list of teams
teams = ["GT", "CSK", "LSG", "MI", "RR", "RCB", "KKR", "PBKS", "DC", "SRH"]

# Initialize dictionaries to store average strike rate and total average for each team
team_avg_sr = {team: 0 for team in teams}
team_avg = {team: 0 for team in teams}

# Iterate through the DataFrame and calculate average strike rate and total average for each team
for team in teams:
    team_data = df_orange[df_orange["TEAM"] == team]

    # Calculate average strike rate
    total_strike_rate = team_data["SR"].astype(float).sum()
    total_batsmen = len(team_data)
    if total_batsmen > 0:
        avg_sr = total_strike_rate / total_batsmen
        team_avg_sr[team] = avg_sr

    # Calculate total average
    team_avg[team] = team_data["AVG"].mean()

# Find the team with the highest strike rate and total average
highest_sr_team = max(team_avg_sr, key=team_avg_sr.get)
highest_avg_team = max(team_avg, key=team_avg.get)

# Plotting the data
fig, ax = plt.subplots(figsize=(12, 6))

# Bar positions
x = range(len(teams))

# Bar width
width = 0.4

# Plot average strike rate
ax.bar([pos - width / 2 for pos in x], [team_avg_sr[team] for team in teams], width, label='Average Strike Rate', alpha=0.7)
# Plot total average
ax.bar([pos + width / 2 for pos in x], [team_avg[team] for team in teams], width, label='Total Average', alpha=0.7)

# Set the x-axis labels to team names
ax.set_xticks(x)
ax.set_xticklabels(teams)

# Add labels and title
ax.set_xlabel('Team')
ax.set_ylabel('Values')
ax.set_title('Average Strike Rate vs Total Average by Team')

# Add legend
ax.legend()

# Annotate the team with the highest strike rate and total average
ax.annotate(f'Highest SR: {highest_sr_team}', xy=(teams.index(highest_sr_team), team_avg_sr[highest_sr_team]), xytext=(-20, 10),
            textcoords='offset points', arrowprops=dict(arrowstyle="->", color='red'))
ax.annotate(f'Highest Avg: {highest_avg_team}', xy=(teams.index(highest_avg_team), team_avg[highest_avg_team]), xytext=(-20, 10),
            textcoords='offset points', arrowprops=dict(arrowstyle="->", color='blue'))

# Show the plot
plt.tight_layout()
plt.show()
```
